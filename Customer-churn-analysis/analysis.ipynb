Customers with lower tenure (new users) have a significantly higher churn rate compared to long-term customers. 
This indicates that churn is highest during the early stages of the customer lifecycle.

### 💡 Recommendation

Focus on improving customer onboarding and engagement in the first few months. 
Providing better support, offers, or guidance early on can help reduce early-stage churn.

### 🔍 Insight: Monthly Charges vs Churn

Customers with higher monthly charges tend to have a higher churn rate compared to those with lower charges.

### 💡 Recommendation

Improve value for high-paying customers by offering better service quality, bundled plans, or loyalty benefits to justify higher charges and reduce churn.

## **Feature Engineering** ##

```python
df['TenureGroup'] = pd.cut(df['tenure'],
                          bins=[0,12,24,48,60,72],
                          labels=['0-1yr','1-2yr','2-4yr','4-5yr','5-6yr'])
```

```python
sns.countplot(x='TenureGroup', hue='Churn', data=df)
plt.show()
```

### 🔍 Insight: Tenure Group vs Churn

Customers in the 0–1 year tenure group have the highest churn rate compared to all other groups. 
This confirms that churn is heavily concentrated in the early stage of the customer lifecycle.

### 🔍 High-Risk Customer Segment

The highest churn is observed among customers who are in their first year, have month-to-month contracts, and pay higher monthly charges. 
These users are early in their journey, lack long-term commitment, and are more sensitive to pricing and service quality.

### 💡 Recommendation

Target customers within their first year by offering onboarding support, personalized offers, and incentives to switch to long-term contracts to reduce early-stage churn.

```python
churn_rate = df.groupby('TenureGroup', observed=False)['Churn'].mean() * 100

churn_rate.plot(kind='bar')
plt.ylabel("Churn Rate (%)")
plt.title("Churn Rate by Tenure Group")
plt.show()
```

### 🔍 Insight: Tenure Group Churn Rate

The 0–1 year tenure group has the highest churn rate at approximately 48%, which is significantly higher than all other tenure groups. 
This confirms that nearly half of the new customers leave within their first year.

### 🔍 Key Finding

Customers most likely to churn are:
- In their first year (0–1 year tenure ~48% churn)
- On month-to-month contracts
- Paying higher monthly charges

This segment represents the highest risk group and should be the primary focus for retention strategies.
